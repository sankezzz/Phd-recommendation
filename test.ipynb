```python
import requests

schools = ["Carnegie Mellon University", "Princeton University", "Yale University", "Columbia University", "University of Washington"]

for school in schools:
    r = requests.get("https://api.openalex.org/institutions", params={
        "search": school,
        "per_page": 1,
        "mailto": "your@email.com"
    })
    result = r.json()["results"][0]
    print(f"{result['display_name']}: {result['id'].split('/')[-1]}")
```
